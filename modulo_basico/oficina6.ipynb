```python
import pandas as pd
import numpy as np
```

```python
dados = {

    'Nome': np.array(['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eduardo']),

    'Idade': np.array([28, 34, 29, None, 42]),

    'Cidade': np.array(['São Paulo', 'Rio de Janeiro', None, 'Curitiba', 'Belo Horizonte']),

    'Vendas': np.array([150, 200, 300, 400, None])

}


df_dados = pd.DataFrame(dados)
df_dados
```

```python
clientes_mais_velhos = df_dados[df_dados['Idade'] > 30]
clientes_mais_velhos
```

```python
media_idade = df_dados['Idade'].mean()
print(f"A média de idade: {media_idade}")
```

```python
df_dados['Cidade'].fillna('Desconhecido', inplace=True)
df_dados
```

```python
soma_vendas = df_dados['Vendas'].sum()
print(f"Soma das vendas: {soma_vendas}")
```
