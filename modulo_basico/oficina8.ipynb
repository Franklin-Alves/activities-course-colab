```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(32)

turma_a = pd.DataFrame({
    "id_aluno": np.arange(0, 30),
    "notas": np.random.choice(["Ótimo", "ruim", "péssimo", "razoável"], size=30),
    "turmas": np.full(30, "8A"),
    "frequencia": np.random.randint(50, 101, size=30)
})

turma_b = pd.DataFrame({
    "id_aluno": np.arange(30, 60),
    "notas": np.random.choice(["A", "B", "C", "F"], size=30),
    "turmas": np.full(30, "8B"),
    "frequencia": np.random.randint(50, 101, size=30)
})

turma_c = pd.DataFrame({
    "id_aluno": np.arange(60, 90),
    "notas": np.random.randint(0, 11, size=30),
    "turmas": np.full(30, "8C"),
    "frequencia": np.random.randint(50, 101, size=30)
})

# Notas Padronizadas
notas_p = {"A": 10, "B": 7, "C":5, "F": 2, "Ótimo": 10, "ruim": 4, "péssimo": 2, "razoável":6}

turmas_juntas = pd.concat([turma_a, turma_b, turma_c])
turmas_juntas["normalizadas"] = turmas_juntas["notas"].replace(notas_p)

turmas_juntas.head()
```

```python
media = turmas_juntas.groupby("turmas")["normalizadas"].mean()
turmas = turmas_juntas["turmas"].unique()

plt.bar(turmas, media.values, color="blue")
plt.title("Comparação das notas por turmas")
plt.xlabel("Tumas")
plt.ylabel("Notas Normalizadas")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="frequencia", y="normalizadas", data=turmas_juntas, hue="turmas")
plt.title("Notas em relação a frequência")
plt.xlabel("Frequência (%)")
plt.ylabel("Notas")
plt.show()
```
