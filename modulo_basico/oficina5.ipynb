```python
import numpy as np
import pandas as pd
```

```python
data = {
    "dias": np.random.choice(["Domingo", "Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sábado"]),
    "horario": np.random.randint(7, 22, size=200),
    "espera": np.random.randint(1, 61, size=200),
    "satisfacao": np.random.randint(1, 11, size=200),
    "funcionarios": np.random.randint(1, 101, size=200),
    "clientes": np.random.randint(1, 151, size=200)
}


df = pd.DataFrame(data)
horarios_data = df["horario"]
turno  = []
for horario in horarios_data:
  if horario >= 7 and horario < 12:
    turno.append("Manhã")
  elif horario < 19:
    turno.append("Tarde")
  else:
    turno.append("Noite")

df['turno'] = turno
```

```python
#Medidas de tendência central

listaMedia = []
listaMediana = []
listaModa = []

turnos_data = df['turno'].unique()

for turno in turnos_data:
  turno_sat = df[df['turno'] == turno]['satisfacao']

  media = turno_sat.mean()
  mediana = turno_sat.median()
  moda = turno_sat.mode().iloc[0]
  listaMedia.append(media)
  listaMediana.append(mediana)
  listaModa.append(moda)

  print(f"Turno: {turno}")
  print()
  print(f"A média de satisfação do turno da {turno}: {media}")
  print(f"A mediana de satisfação do turno da {turno}: {mediana}")
  print(f"A moda de satisfação turno da {turno}: {moda}")
  print()

df_tendencia = pd.DataFrame({
      "Turno": turnos_data,
      "Média": listaMedia,
      "Mediana": listaMediana,
      "Moda":listaModa
})

df_tendencia
```

```python
#Medidas de dispersão
listaAmplitude = []
listaVariancia = []
listaDesvio = []
listaCoeficiente = []

for turno in turnos_data:
  turno_sat = df[df['turno'] == turno]['satisfacao']
  amplitude = turno_sat.max() - turno_sat.min()
  variancia = turno_sat.var()
  desvio_padrao = turno_sat.std()
  coeficiente_variacao = desvio_padrao / turno_sat.mean() if turno_sat.mean() != 0 else 0
  listaAmplitude.append(amplitude)
  listaVariancia.append(variancia)
  listaDesvio.append(desvio_padrao)
  listaCoeficiente.append(coeficiente_variacao)

  print(f"Turno {turno}:")
  print()
  print(f"A Amplitude do turno da {turno} é: {amplitude}")
  print(f"A Variância do turno da {turno} é: {variancia}")
  print(f"O Desvio Padrão do turno da {turno} é: {desvio_padrao}")
  print(f"O Coeficiente de Variação do turno da {turno} é: {coeficiente_variacao}")
  print()

df_dispersao = pd.DataFrame({
      "Turno": turnos_data,
      "Amplitude": listaAmplitude,
      "Variância": listaVariancia,
      "Desvio Padrão": listaDesvio,
      "Coeficiente de Variação": listaCoeficiente
})

df_dispersao
```

```python
tot_turno_m = len(df[df['turno'] == "Manhã"])
tot_turno_t = len(df[df['turno'] == "Tarde"])
tot_turno_n = len(df[df['turno'] == "Noite"])

turno_m = len(df[(df['turno'] == "Manhã") & (df['satisfacao'] > 5)])
turno_t = len(df[(df['turno'] == "Tarde") & (df['satisfacao'] <= 4)])
turno_n = len(df[(df['turno'] == "Noite") & (df['satisfacao'] == 8)])

prob_m = turno_m / tot_turno_m
prob_t = turno_t / tot_turno_t
prob_n = turno_n / tot_turno_n

print(f"A probabilidade dos clientes do turno da manhã terem uma satisfação maior que 5 é: {prob_m:.2f}")
print(f"A probabilidade dos clientes do turno da tarde terem uma satisfação menor ou igual 4 é: {prob_t:.2f}")
print(f"A probabilidade dos clientes do turno da noite terem uma satisfação igual a 8 é: {prob_n:.2f}")
```

```python
#Intervalo de confiança

from scipy import stats
import numpy as np

def intervalo_confianca(media, desvio_padrao, n, nivel=0.95):
  z = stats.norm.ppf(1 - (1 - nivel) / 2)
  margem_de_erro = z * (desvio_padrao / np.sqrt(n))
  return media - margem_de_erro, media + margem_de_erro

resultados_ic = []

for turno in sorted(df['turno'].unique()):
  turno_sat = df[df['turno'] == turno]['satisfacao']
  media = turno_sat.mean()
  desvio_padrao = turno_sat.std()
  amostra = len(turno_sat)
  ic_inferior, ic_superior = intervalo_confianca(media, desvio_padrao, amostra)

  resultados_ic.append({
      "Turno": turno,
      "Intervalor de confiança inferior": ic_inferior,
      "Intervalor de confiança superior": ic_superior
  })

df_ic = pd.DataFrame(resultados_ic)
df_ic
```

```python
#Hipóteses

satisfacao_m = df[df['turno'] == "Manhã"]["satisfacao"]
satisfacao_t = df[df['turno'] == "Tarde"]['satisfacao']

#Realizando teste de t
t_stat, p_value = stats.ttest_ind(satisfacao_m, satisfacao_t)

print(f"Estátisca t: {t_stat:.2f}")
print(f"O valor de p: {p_value:.2f}")
```
